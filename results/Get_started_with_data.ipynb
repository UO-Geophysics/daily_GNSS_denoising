# quick guide on how to interact with the denoised data

```python
import xarray as xr
import matplotlib.pyplot as plt
```

## open the dataset

```python
path = "../data/results/UNR_GNN_2010_2023.nc"
ds = xr.open_dataset(path)
ds
```

## Select a station and time

```python
station = "ALBH"
ds_ALBH = ds.sel(station=station)
ds_ALBH
```

```python
ds_ALBH_SSE = ds_ALBH.sel(time=slice("2017-03-01", "2017-05-01"))
ds_ALBH_SSE
```

## Simple plot

```python
ds_ALBH_SSE.original_e.plot(label="Original")
ds_ALBH_SSE.final_denoised_e.plot(label="Denoised")
plt.legend()
```
